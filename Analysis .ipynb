```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('results.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df_pivot = df.pivot(index='date', columns='name', values='average_volatility')
plt.figure(figsize=(14,8))
for column in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[column], label=column)

plt.title('Average Volatility Trend per Company')
plt.xlabel('Date')
plt.ylabel('Average Volatility')
plt.legend()
plt.show()
```

```python
exclude_date = pd.to_datetime('2023-05-19')
df_pivot = df_pivot[df_pivot.index != exclude_date]
ax = df_pivot.plot(kind='bar', stacked=False, figsize=(14,8))

ax.set_xticklabels([x.strftime('%Y-%m-%d') for x in df_pivot.index])

plt.xticks(rotation=0)

plt.title('Daily Highest Volatility per Company')
plt.xlabel('Date')
plt.ylabel('Highest Volatility')
plt.legend()
plt.tight_layout()
plt.show()
```

***The relationship between two visualization makes sense. If a company shows high daily volatility, it's likely to have a high average volatility as well.However, there can be some small variations due to the case of extreme volatility, which can skew the average***
